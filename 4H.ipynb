```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
import seaborn as sns
```

```python
df=pd.read_csv('/content/drive/MyDrive/DSBDA/heartdisease.csv')
```

```python
df
```

```python
df.columns
```

```python
df1=df.loc[1:5]
```

```python
x=df1['Age']
y=df1['ChestPain']
mpl.plot(x,y)
mpl.title("Line Chart")
mpl.xlabel("Age")
mpl.ylabel("ChestPain")
mpl.show()
```

```python
x=df['Age']
mpl.hist(x)
mpl.title("Histogram")
mpl.xlabel("Age")
mpl.show()
```

```python
x=df['Sex']
y=df['Ca']
mpl.bar(x,y)
mpl.title("Bar Chart")
mpl.xlabel("Age")
mpl.ylabel("Ca")
mpl.show()
```

```python
x=df['Age']
y=df['Chol']
mpl.scatter(x,y)
mpl.title("Scatter Plot")
mpl.xlabel("Age")
mpl.ylabel('Chol')
mpl.show()
```

```python
x=df['Sex'].value_counts()
y=['Male','Female']
mpl.pie(x,labels=y)
mpl.title("Pie Chart")
mpl.show()
```
