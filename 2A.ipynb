```python
df['AQI_Bucket'].unique()
```

```python
# meanval=df['AQI_Bucket'].mean()
df['AQI_Bucket']=df['AQI_Bucket'].replace('Poor',2)
df['AQI_Bucket']=df['AQI_Bucket'].replace('Very Poor',1)
df['AQI_Bucket']=df['AQI_Bucket'].replace('Satisfactory',3)
df['AQI_Bucket']=df['AQI_Bucket'].replace('Severe',4)
df['AQI_Bucket']=df['AQI_Bucket'].replace('Moderate',5)
df['AQI_Bucket']=df['AQI_Bucket'].replace('Good',6)
```

```python
meanval = df['AQI_Bucket'].astype(int).mean()
df['AQI_Bucket'] = df['AQI_Bucket'].replace(0, meanval)
```

```python
df['AQI_Bucket'].unique()
```

```python
# prompt: build a model for this dataset

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X = df[['PM2.5', 'PM10', 'NO', 'NO2', 'NOx']]
y = df['AQI_Bucket']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
from sklearn.metrics import mean_squared_error, r2_score
print('Mean squared error:', mean_squared_error(y_test, y_pred))
print('R2 score:', r2_score(y_test, y_pred))
```
